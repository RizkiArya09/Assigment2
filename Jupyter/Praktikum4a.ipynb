```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("employee.csv")
dataset
```

```python
#Get the age attribute's value
dage = np.array(dataset)
dage = dage[:, [0,2]]
print ("Index 0 = ID, Index 1 = Age")
pd.DataFrame(dage)
```

```python
age01=np.sum(dage[0:1,1],axis=0)
age02=np.sum(dage[1:2,1],axis=0)
age03=np.sum(dage[2:3,1],axis=0)
age04=np.sum(dage[3:4,1],axis=0)
age05=np.sum(dage[4:5,1],axis=0)
print("age 01:",age01)
print("age 02:",age02)
print("age 03:",age03)
print("age 04:",age04)
print("age 05:",age05)
```
